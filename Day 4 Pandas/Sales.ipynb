```python
import numpy as np
data=np.array([10,12,14,15,18,21,25,100])
Q1=np.percentile(data,25)
Q3=np.percentile(data,75)
IQR=Q3-Q1
lower_bound=Q1-1.5 * IQR
upper_bound=Q3 + 1.5 * IQR
outliers=data[(data<lower_bound) | (data>upper_bound)]
print("Outliers:",outliers)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
#create the data
sales=np.array([1300,1500,1500,1700,1900,2000,1750,2100,1575,1450])
days=np.arange(1,11) # array of 1 to 10
```

```python
print(sales)
```

```python
print(days)
```

```python
print(max(sales))
```

```python
print(min(sales))
```

```python
#basic analysis
total_sales=np.sum(sales)
average_sales=np.mean(sales)
highest_sales=np.max(sales)
lowest_sales=np.min(sales)
print("Total Sales", total_sales)
print("Average Sales", average_sales)
print("highest Sales", highest_sales)
print("Lowest Sales", lowest_sales)
```

```python
#sales trends over a days
plt.figure(figsize=(8, 5))
plt.plot(days,sales,marker='o',linestyle='-',color='b',label="sales")
plt.axhline(average_sales,color='r',linestyle='--',label="average_sales")
plt.xlabel("Days")
plt.ylabel("Sales Revenue ($)")
plt.title("Sales Trend Over 10 Days")
plt.legend()
plt.grid()
plt.show()
```

```python
#sales comparision by days
plt.figure(figsize=(8, 5))
plt.bar(days,sales,color='skyblue',label="sales")
plt.xlabel("Days")
plt.ylabel("Sales Revenue ($)")
plt.title("Sales per Day")
plt.legend()
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
#pie chart
plt.figure(figsize=(6, 6))
plt.pie(sales,labels=days,colors=plt.cm.Paired.colors,autopct='%1.1f%%')
plt.title("Sales Contribution Per Day")
plt.show()
```
